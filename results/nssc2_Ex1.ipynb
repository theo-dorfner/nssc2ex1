```python
import numpy as np
import matplotlib.pyplot as plt
iterations = 30
resolutions = [125,250,1000,2000]
processes = [1,2,3,5,7,10,12,15,17,20,22,25,27,30,32,35,37,40]
```

# Collect Data

```python
serialTime = [2.092500e+00 / 2000, 8.642756e+00 / 2000, 1.430785e+02 /2000, 5.541223e+02 /2000] # this is for the four resolutions
print(serialTime)
```

```python
# resolution 125
file125 = open("stdout-res125.txt",'r')
fullResults125 = []
run = []
for line in file125:
    if("jacobiMPI" in line): 
        continue
    if("---" in line):
        fullResults125.append(run)
        run = []
        continue
    num = float(line.split(" ")[1])
    run.append(num)
print(fullResults125)
```

```python
# resolution 250
file250 = open("stdout-res250.txt",'r')
fullResults250 = []
run = []
for line in file250:
    if("jacobiMPI" in line): 
        continue
    if("---" in line):
        fullResults250.append(run)
        run = []
        continue
    num = float(line.split(" ")[1])
    run.append(num)
print(fullResults250)
```

```python
# resolution 1000
file1000 = open("stdout-res1000.txt",'r')
fullResults1000 = []
run = []
for line in file1000:
    if("jacobiMPI" in line): 
        continue
    if("---" in line):
        fullResults1000.append(run)
        run = []
        continue
    num = float(line.split(" ")[1])
    run.append(num)
print(fullResults1000)
```

```python
# resolution 2000
file2000 = open("stdout-res2000.txt",'r')
fullResults2000 = []
run = []
for line in file2000:
    if("jacobiMPI" in line): 
        continue
    if("---" in line):
        fullResults2000.append(run)
        run = []
        continue
    num = float(line.split(" ")[1])
    run.append(num)
print(fullResults2000)
```

# Extract Time Values

```python
times125 = []
times250 = []
times1000 = []
times2000 = []
for i in range(len(processes)):
    times125.append(fullResults125[i][-1])
    times250.append(fullResults250[i][-1])
    times1000.append(fullResults1000[i][-1])
    times2000.append(fullResults2000[i][-1])
```

```python
timeOverview = [times125,times250,times1000,times2000]
parallelSpeedup = []
parallelEfficiency = []
for i,res in enumerate(timeOverview):
    runSpeed = []
    runEff = []
    for j,time in enumerate(res):
        sTime = serialTime[i]
        runSpeed.append(sTime/time)
        runEff.append(sTime/(time*processes[j]))
    parallelSpeedup.append(runSpeed)
    parallelEfficiency.append(runEff)
```

```python
print(parallelEfficiency)
```

# Plotting

```python
# parallel speedup
for i in range(4):
   plt.semilogy(processes,parallelSpeedup[i],'o-',label="Resolution = " + str(resolutions[i]))
#plt.plot(processes,parallelSpeedup[3],'o-',label="Resolution = " + str(resolutions[i]))
plt.legend()
plt.title("Parallel Speedup for MPI Parallelization")
plt.xlabel("Number of Processes")
plt.ylabel("Parallel Speedup")
#plt.savefig("parallelSpeedup.png",dpi=300)
```

```python
# parallel speedup
for i in range(4):
    plt.semilogy(processes,parallelEfficiency[i],'o-',label="Resolution = " + str(resolutions[i]))
plt.legend()
plt.title("Parallel Efficiency for MPI Parallelization")
plt.xlabel("Number of Processes")
plt.ylabel("Parallel Efficiency")
#plt.savefig("parallelEfficiency.png",dpi=300)
```
